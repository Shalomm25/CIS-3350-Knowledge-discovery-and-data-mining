```python
import pandas as pd 

myData = pd.read_csv('titanic.csv')


# survival ---> Survival 
# pclass ---> Ticket class
# sex ---> Gender 
# age ---> age in years
# sibsp ---> # of siblings / spouses aboard the titanic
# parch ---> # of parents / childrren aboard the Titanic 
# ticket ---> Ticket number
# cabin ---> Cabin Number 
# embaked ---> Port of Embarkation


myData.head(10)
```

```python
# find out how many missing datapoints are there in a dataset
myData.isna().sum()
# by adding the sum function to the isna function we get to see how many missing values for each variable
```

```python
# drop rows with missing data 
myData.dropna()
# we get an empty data set because it dropped every row because each row had 1 or more missing values
# myData.dropna().shape # when adding the shape function to the dropna function we can see that there are 0 rows and 14 columns (variables)
```

```python
# drop columns with missing data 
myData.dropna(axis = 'columns')
# we still have rows but none of the columns that have missing values
# myData.dropna(axis = 'columns').shape #by adding the shape function to the dropna function we are recognzing that all the 1309 rows are there but now only 7 columns remain
```

```python
# data imputation - mean (for nemerical data)

myData.fillna(myData.mean()) # The mean function for all numeric data will plug in the mean of all the numeric data 
```

```python
myData.mean() # the mean is the average
```

```python
# data imputation - median (for numerical data)

myData.fillna(myData.median()) 
```

```python
myData.median() # the median is the middle number in ascending or descending order in a list of numbers
```

```python
# fill in custom data

myData['boat'] = myData['boat'].fillna('noBoat') # we are taking the missing data in the 'boat' column 'NaN' and replacing it with the value 'noBoat'

myData.head(20)
```
