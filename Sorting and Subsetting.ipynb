```python
import pandas as pd

myData = pd.read_csv('titanic.csv') # csv files

myData.head(8)
```

```python
# number of women on board

(myData['sex'] == 'female').sum()
# myData.sex == 'female'
```

```python
# number of woman on board who survived 

((myData['sex'] == 'female') & (myData['survived'] == 1)).sum()
```

```python
# sort data by fare paid

# myData.sort_values('fare', ascending = False)

# sortedvaluesbyfare = myData.sort_values('fare', ascending = False)

# sortedvaluesbyfare.head(10)



myData.sort_values('fare', ascending = False) # for the second argument by making ascending = False it automaytically knows to sort in descending order
```

```python
#  sort by fare paud and age
myData.sort_values(['fare', 'age'], ascending = [False, True]) # this helps create a clear order between two or more variables
# it is automatically in ascending order
```

```python
# subset dataset

women_on_titanic = myData.query("sex == 'female' ")

women_on_titanic
```

```python
# subset dataset by several charachteristic

myData.query("sex == 'female' & (embarked == 'S' | embarked == 'C' )") # this allows you to get even more specific information using two or more variables
```

```python
# filter for under 30 years old

myData.query("age <= 30 ")
```

```python
# filter for over 60 years old male

myData.query("age > 60 & sex == 'male' ").sort_values('fare', ascending=False)
```

```python
# save datafile to computer

#  exclude the "index" column
saved_data = myData.query("age > 60 & sex == 'male' ").sort_values('fare', ascending=False)

saved_data.to_csv('datasaved.csv')
```
